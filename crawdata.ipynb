```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import csv
import re
import time
import pandas as pd
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
driver = webdriver.Chrome(chrome_options=chrome_options,executable_path='C:/Users/BUIQUOCHUY/Desktop/CODE-STUDY/BA/Stock-Prediction/Bot_Craw_data/chromedriver_win32/chromedriver')
link_base='https://hoclaixemoto.com/phan-kien-thuc-luat-200-cau-hoi/'
```

```python

driver.get(link_base)
data=[]
count=driver.find_elements_by_class_name('accordion-title')
data
for i in range(10,((len(count)*7)+9),7):
   while True:
      try:
         question=driver.find_element_by_css_selector(f"#post-2226 > div > div > p:nth-child({i}) > strong")
         
         try:
            a=driver.find_element_by_css_selector(f"#post-2226 > div > div > p:nth-child({i}) + p")
            answer1=a.text
         except:
            answer1=""
         try:
            b=driver.find_element_by_css_selector(f"#post-2226 > div > div > p:nth-child({i}) + p + p")
            answer2=b.text
         except:
            answer2=""
             
         try:
            c=driver.find_element_by_css_selector(f"#post-2226 > div > div > p:nth-child({i}) + p + p + p")
            answer3=c.text
         except:
            answer3=""
         try:
            d=driver.find_element_by_css_selector(f"#post-2226 > div > div > p:nth-child({i}) + p + p + p + p")
            answer4=d.text
         except:
            answer4=""
         break
      except:
         i=i+1
   data.append([question.text,answer1,answer2,answer3,answer4])
df = pd.DataFrame(data,columns=['question','answer1','answer2','answer3','answer4'])
print(df)
df.to_excel("outputA2.xlsx")
```
